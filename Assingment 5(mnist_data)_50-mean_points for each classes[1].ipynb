```python
import pandas as pd
import numpy as np
import random 
```

```python
df1=pd.read_csv("D:\Data_Analytaics\Sem-3\Basics of ML(DC-302)\Data\mnist_train.csv")
df2=pd.read_csv("D:\Data_Analytaics\Sem-3\Basics of ML(DC-302)\Data\mnist_test.csv")
```

```python
X_train=np.array(df1)
X_train.shape
```

```python
X_test=np.array(df2.loc[:,df2.columns!='label'])
Y_test=np.array(df2['label'])
```

```python
len(X_test),len(Y_test)
```

```python
#classes
classes=[]
for i in range(10):
    classes.append(X_train[X_train[:,0] == i])
classes   
```

```python
#duplicate class  
class1 = []
for i in range(10):
    class1.append(X_train[X_train[:,0] == i])
class1[0]
```

```python
sum=0
for i in classes:
    print(len(i))
    sum+=len(i)
print(sum)
```

```python
sample1=[]
for i in range(10):
    for j in range(5):
        sample1.append(np.array(random.sample(list(class1[i]),int(len(class1[i])//5))))
        class1[i] = classes[i]
sample1
```

```python
mean_dist={}
for i in range(len(classes)):
    mean_dist[f'{i}'] = np.mean(sample1[i][:,1:],axis=0)
mean_dist 
```

```python
mean_dist={}
for i in range(50):
    mean_dist[f'{i}']=np.mean(sample1[i][:,1:],axis=0)
mean_dist
```

```python
lst = []
for i in range(10):
    for j in range(5):
        lst.append(i)
np.array(lst)
```

```python

Y_pred = []
for i in range(10000):
    
    distance = []
    index = 0
    Y = []
    for j in range(10):
        for k in range(5):
            distance.append(np.linalg.norm(X_test[i] - mean_dist[f'{index}']))
            index += 1
    distance = np.array(distance)
    distance = np.stack((distance,lst),axis = 1)
    distance = distance.astype("int64")
    distance = distance[distance[:,0].argsort()]
    Y = distance[:5]
    Y
    var = 0
    
    for j in range(5):
        for n in Y[:,1]:
            if np.count_nonzero(Y[:,1]== n) > 1:
                var = n
            else:
                var = n
    Y_pred.append(n)
```

```python
Y_pred
```

```python
count=0
for i in range(10000):
    if Y_pred[i]==Y_test[i]:
        count+=1
count
```

```python
Total_accuracy= count/len(Y_test)*100
print("total accuracy: ",Total_accuracy,'%')
```

```python
#class_wise comparision
compare=np.stack((Y_pred,Y_test),axis = 1)
compare
```

```python
f_count=[]
for i in range(len(compare)):
    if compare[i,0] != compare[i,1]:
        f_count.append(Y_test[i])
    
len(f_count)
    
```

```python
for i in set(Y_pred):
    count_=list(Y_test).count(i)
    freq=f_count.count(i)
    print(f'Accuracy for class{i}:',100-freq/count_*100 )
```
