```python

a = []
for i in range(10000):
    
    dist = []
    index = 0
    X = []
    for j in range(10):
        for k in range(5):
            dist.append(np.linalg.norm(test1[i] - mean[f'{index}']))
            index += 1
    dist = np.array(dist)
    dist = np.stack((dist,one),axis = 1)
    dist = dist.astype("int64")
    dist = dist[dist[:,0].argsort()]
    X = dist[:5]
    X
    var = 0
    
    for j in range(5):
        for n in X[:,1]:
            if np.count_nonzero(X[:,1]== n) > 1:
                var = n
            else:
                var = n
    a.append(n)
```
