```python
import pandas as pd
import numpy as np
dataset = pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\iris dataset')
dataset.head()
```

```python
new_columns_name = []
for i in range(1,len(dataset.columns[:4])+1):
    name = 'X'
    new_columns_name.append(name+str(i))
new_columns_name.append('Y')
dataset.columns = new_columns_name
print(dataset.columns)
dataset
```

```python
dataset['Y'].unique()
```

```python
number = dataset['Y'].unique()
values = [i for i in range(len(number))]
mapp = dict(zip(number,values))
print(mapp)
```

```python
dataset['Y'] = dataset['Y'].replace(number,values)
dataset
```

```python
features = dataset.to_numpy()
print(features)
```

```python
clusters = {}
for name in mapp.keys():
    clusters[name]=features[np.where(features[:,4]==mapp[name])]
print(clusters)
```

```python
x = np.empty((0,5))
y = np.empty((0,5))
for class_ in clusters.keys():
    x = np.array(np.append(x,clusters[class_][:-10],axis=0))
    y = np.array(np.append(y,clusters[class_][:-10],axis=0))
training_dataset=x
testing_dataset=y
```

```python
_class_ = {}
for name in mapp.keys():
    _class_[name]=training_dataset[np.where(training_dataset[:,4]==mapp[name])][:,:4]
_class_
```

```python
cluster_mean = dict(zip([f"mean{X}" for X in range(1,len(_class_.keys())+1)],[np.mean(_class_[name],axis=0) for name in _class_.keys()]))
cluster_mean
```

```python
def testing(test_points):
    dist1=np.sqrt(np.sum(np.square(np.subtract(cluster_mean['mean1'],test_points))))
    dist2=np.sqrt(np.sum(np.square(np.subtract(cluster_mean['mean2'],test_points))))
    dist3=np.sqrt(np.sum(np.square(np.subtract(cluster_mean['mean3'],test_points))))
    if dist1<=dist2 and dist1<=dist3:
        cluster = 0
    elif dist2<=dist3 and dist2<=dist1:
        cluster = 1
    else:
        cluster = 2
    return cluster
```

```python
correct = 0
incorrect = 0
for i in testing_dataset[:,:5]:
    count = 1
    test_points = i[:4]
    if testing(test_points)==i[4]:
        correct+=1
    else:
        incorrect+=1
print('correct result: ',correct,':incorrect result: ',incorrect)
accuracy = (correct/len(testing_dataset))*100
print('Accuracy is: ',accuracy)
```
