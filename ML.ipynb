```python
import random
cluster1 = []
for i in range(100):
    cluster1.append([random.uniform(2,4),random.uniform(2,4)])
                     
cluster2 = []
for i in  range(100):
    cluster2.append([random.uniform(0,2),random.uniform(0,2)])
        
import matplotlib.pyplot as plt
plt.scatter([i[0] for i in cluster1],[i[1] for i in cluster1])
plt.scatter([i[0] for i in cluster2],[i[1] for i in cluster2])
mean1 = [sum(i[0] for i in cluster1)/100,sum(i[1] for i in cluster1)/100]
mean2 = [sum(i[0] for i in cluster2)/100,sum(i[1] for i in cluster2)/100]
plt.scatter(mean1[0],mean1[1])
plt.scatter(mean2[0],mean2[1])

x,y = random.uniform(0,4),random.uniform(0,4)
plt.scatter(x,y)

def test(x,y):
    dist1 = ((mean1[0]-x)**2+(mean1[1]-y)**2)**0.5
    dist2 = ((mean2[0]-x)**2+(mean2[1]-y)**2)**0.5
    if dist1<=dist2:
        print("BELONGS TO CLUSTER1")
    else:
        print("BELONGS TO CLUSTER2")
test(x,y)
```
