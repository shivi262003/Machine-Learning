```python
#Assignment 2
import numpy as np
cluster1 = np.random.uniform(0,2, size = (100,1000))
cluster2 = np.random.uniform(2,4, size = (100,1000))

mean_cluster1 = np.mean(cluster1 , axis=0)
mean_cluster2 = np.mean(cluster2 , axis=0)
print(mean_cluster1)
print(mean_cluster2)
def test(test_points):
    dist1 = np.sqrt(np.sum(np.square(np.subtract(mean_cluster1,test_points))))
    dist2 = np.sqrt(np.sum(np.square(np.subtract(mean_cluster2,test_points))))
    print(dist1)
    print(dist2)
    if dist1>dist2:
        print("Belongs to cluster1")
    else:
        print("Belongs to cluster2")
```

```python
test_points=np.random.uniform(0,4,size=1000)
print(test_points)
```

```python
test(test_points)
```
